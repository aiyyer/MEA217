```python
import pandas as pd
```

```python
input_file = './co2_daily_mlo.csv'

# we need to skip the first 32 lines of the data description
# This data has no headers. We will create them as we read
#
df = pd.read_csv(input_file, skiprows=32, header=None,
                 names=['year', 'month', 'day', 'frac', 'conc' ])
```

```python
# inspect the dataframe
df.head()
```

```python
# rename columns
df.columns = ['Year', 'mon', 'day', 'frac_year', 'data']
```

```python
# what is the long term average of co2?
```

```python
co2_ave = df['data'].mean()
print("Long term mean = ", co2_ave)
```

```python
# min and max values of co2

co2_min = df['data'].min()
co2_max = df['data'].max()

print(co2_min, co2_max)
```

```python
# when did the min occur?
matching_rows = df.loc[df['data'] == df['data'].min()]
print(matching_rows)
```

```python
# when did the max occur?
matching_rows = df.loc[df['data'] == df['data'].max()]
print(matching_rows)
```

```python
# lets average data over each year 
yearly_avg = df.groupby('Year')['data'].mean()
yearly_avg.plot()
```

```python
df['data'].plot()
```

```python
# Annual cycle of co2
ann_cycle = df.groupby('mon')['data'].mean()

ann_cycle.plot(marker='o')
```
